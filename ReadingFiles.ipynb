```python
file = open("spider.txt")
print(file.readline())   #file will not be found (Just an example) Reads first line in a file
```

```python
print(file.readline())   #will read the next line in the file
```

```python
print(file.read())       #Starts from where file.readline left off, but reads the rest of the lines
```

```python
file.close()             #close the file after use
```

```python
with open("spider.txt") as file:
  print(file.readline())  #If you use a with block, the file will close after use
```
