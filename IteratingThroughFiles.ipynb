```python
with open("spider.txt") as file:
  for line in file:
    print(line.upper())         #Prints everything in upper case (New line charachter creates doublespacing)
```

```python
with open("spider.txt") as file:
  for line in file:
    print(line.strip().upper())         #Prints everything in upper case (.strip() deletes new line character)
```

```python
file = open(spider.txt)
lines = file.readlines()
file.close()                            #lines variable keeps lines in the file
lines.sort()
print(lines)
```
