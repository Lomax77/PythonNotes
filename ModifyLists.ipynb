```python
fruits = ["Pineapple", "Banana", "Apple", "Melon"]
fruits.append("Kiwi")
print(fruits)
```

```python
fruits.insert(0, "Orange")
print(fruits)
```

```python
fruits.insert(25, "Peach")
print(fruits)
```

```python
fruits.remove("Melon")
print(fruits)
```

```python
fruits.pop(3) #Removes the 4th word in the list
```

```python
print(fruits) #'Apple' is now removed
```

```python
fruits[2] = "Strawberry" #"Strawberry added to 3rd spot in list"
print(fruits)
```
